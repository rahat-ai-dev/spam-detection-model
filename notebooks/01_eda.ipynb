```python
#Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
df = pd.read_csv(r'C:\spam-detection\data\raw\sms_spam_dataset.csv', encoding='latin-1')
df.columns = ['label', 'message']
df.head()
```

```python
#Print how many spam and how many ham
df['label'].value_counts()
```

```python
#Show Spam vs Ham through chart
df['label'].value_counts().plot(kind='bar',color=['green','red'])
plt.title('Spam vs Ham')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
#print the length of message
df['length']=df['message'].apply(len)
df.groupby('label')['length'].mean()
```

```python
#Show length distribution
df.hist(column='length',by='label', bins=50, figsize=(10,4))
plt.show()
```
