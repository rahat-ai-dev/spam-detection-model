```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import os
import scipy.sparse as sp
```

```python
#Load the processed data

if os.getcwd().endswith('notebooks'):
    os.chdir('..')

train_df=pd.read_csv(os.path.join('data','processed','train.csv'))
test_df=pd.read_csv(os.path.join('data','processed','test.csv')) 

train_df['message'] = train_df['message'].fillna("")
test_df['message']=test_df['message'].fillna("")
   
x_train=train_df['message']
y_train=train_df['label']

x_test=test_df['message']
y_test=test_df['label']

print(f'Train Size: {len(x_train)}')
print(f'Test size:{len(x_test)}')
```

```python
# Make IF-IDF

vectorizer=TfidfVectorizer(max_features=3000)

# Train the model with training data
x_train_tfidf = vectorizer.fit_transform(x_train)

#Only transform the test data
x_test_tfidf = vectorizer.transform(x_test)

print(f'Feature metrix shape:{x_train_tfidf.shape}')



```

```python
# See what features were made
feature_name = vectorizer.get_feature_names_out()
print(f'Total features: {len(feature_name)}')
print(f'Sample feature: {feature_name[:20]}')
```

```python
#Save the Vectoriger
pickle.dump(vectorizer, open('models/tfidf_vectorizer.pkl', 'wb'))
print('Vectorizer saved!')
```

```python
#Save the feature matrix
sp.save_npz('data/processed/x_train_tfidf.npz', x_train_tfidf)
sp.save_npz('data/processed/x_test_tfidf.npz',x_test_tfidf)

y_train.to_csv('data/processed/y_train.csv', index=False)
y_test.to_csv('data/processed/y_test.csv',index=False)

print('Features saved!')
```
