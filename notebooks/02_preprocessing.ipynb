```python
#Import library
import pandas as pd
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split
import os
```

```python
#Load the data
df = pd.read_csv(r'C:\spam-detection\data\raw\sms_spam_dataset.csv', encoding='latin-1')
df.columns = ['label', 'message']
df.head()
```

```python
#Function to clean the text
stemmer= PorterStemmer()
stop_words=set(stopwords.words('english'))

def clean_text(text):
    #Make lowercase
    text=text.lower()
    #Remove number and punctuation
    text=re.sub(r'[^a-z\s]','',text)
    #Seperate words
    words=text.split()
    #Remove stopwords and do stemming
    words=[stemmer.stem(w) for w in words if w not in stop_words]
    return ' '.join(words)
 
```

```python
# Test function with an example
test = "FREE entry! Win a cash prize! Click now!!!"
print(clean_text(test))
```

```python
#Apply into full dataset
df['clean_message']=df['message'].apply(clean_text)
df[['message','clean_message']].head()
```

```python
#Convert lavel into number
# spam = 1, ham = 0
df['label_num']=df['label'].map({'spam':1, 'ham':0})
df.head()
```

```python
# Do train test split
x=df['clean_message']
y=df['label_num']
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)

print(f'Train size:{len(x_train)}')
print(f'Test size:{len(x_test)}')
```

```python
# Save the processed data

if os.getcwd().endswith('notebooks'):
    os.chdir('..')

output_dir=os.path.join('data','processed')

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

train_df=pd.DataFrame({'message': x_train, 'label':y_train})
test_df=pd.DataFrame({'message':x_test, 'label':y_test})

train_df.to_csv(os.path.join(output_dir,'train.csv'),index=False)
test_df.to_csv(os.path.join(output_dir,'test.csv'),index=False)

print('Data saved!')
```
