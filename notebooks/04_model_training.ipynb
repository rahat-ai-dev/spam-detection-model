```python
import pandas as pd
import numpy as np
import os
import scipy.sparse as sp
import pickle
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the Data

if os.getcwd().endswith('notebooks'):
    os.chdir('..')

x_train=sp.load_npz(os.path.join('data','processed','x_train_tfidf.npz'))
x_test=sp.load_npz(os.path.join('data','processed','x_test_tfidf.npz'))

y_train=pd.read_csv(os.path.join('data','processed','y_train.csv'))['label']
y_test=pd.read_csv(os.path.join('data','processed','y_test.csv'))['label']

print(f'Train size:{x_train.shape}')
print(f'test size:{x_test.shape}')
```

```python
# Train 3 Models

#Naive Bayes
nb=MultinomialNB()
nb.fit(x_train, y_train)

#Logistic Regression
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)

#Random Forest
rf=RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train,y_train)

print('All models are trained!')
```

```python
#Compare the 3 models
models={
    'Naive Bayes': nb,
    'Logistic Regression': lr,
    'Random Forest': rf
    }
for name,model in models.items():
    acc= accuracy_score(y_test,model.predict(x_test))
    print(f'{name}:{acc:.4f}')
    
```

```python
#Save the best model
best_model=lr
pickle.dump(best_model,open('models/spam_classifier.pkl','wb'))
print('Best model saved!')
```
